## Аутентификация

Код взят из <a href='https://github.com/python273/vk_api/blob/6707d740ab069015ae76f3cfa1449ebe55d0a35c/examples/two_factor_auth.py'>примера использования vk_api</a>. Для подключения функций API сначала необходимо выполнить вход в аккаунт.

```python
def auth_handler():
    """ При двухфакторной аутентификации вызывается эта функция.
    """

    # Код двухфакторной аутентификации
    key = input("Enter authentication code: ")
    # Если: True - сохранить, False - не сохранять.
    remember_device = True

    return key, remember_device
```

```python
#обычный вход
#vk_session = vk_api.VkApi(login, password)

#двухфакторная аутентификация
vk_session = vk_api.VkApi(login, 
                          password,# функция для обработки двухфакторной аутентификации
                          auth_handler=auth_handler,
                         )
vk_session.auth(token_only=True)
vk = vk_session.get_api()
```

## Получим список друзей и друзей друзей от VK API

Получим список друзей текущего пользователя. Для этого воспользуемся функцией API <a href='https://vk.com/dev/friends.get'>friends.get</a>. Добавим парметр *order='hints'* для получения списка друзей в том же порядке, в котором они отсортированы в разделе *Мои друзья*.

```python
my_friends = vk.friends.get(fields=['nickname', 'domain', 'sex', 'bdate', 'city',
                                    'country', 'timezone', 'photo_50', 'photo_100', 
                                    'photo_200_orig', 'has_mobile', 'contacts', 
                                    'education', 'online', 'relation', 'last_seen',
                                    'status', 'can_write_private_message', 
                                    'can_see_all_posts', 'can_post', 'universities'],
                            order='hints'
                           )
print('Количество друзей:', my_friends['count'])
```

Сохраним на диск

```python
#save pkl
with open('my_friends.pkl', 'wb') as fp:
    pickle.dump(my_friends, fp, protocol=pickle.HIGHEST_PROTOCOL)
    
'''
#load pkl
with open('my_friends.pkl', 'rb') as fp:
    data = pickle.load(fp)'''
```

VK API на любой запрос возвращает json-файл, который в питоне представляется как словарь (dict). Он имеет два элемента с ключами:

- count - число элементов в ответе API
- items - список ответов API. Каждый элемент в ответе - это словарь

```python
print(type(my_friends))

pprint(my_friends['items'][:3])
```

Теперь получим друзей друзей. Воспользуемся *try except* для обработки исключений, которые возникают при попытке получить друзей заблокированного или уделенного пользователя. Добавим задержку, т.к. при обнаружении потока обращений к одной и той же функции API аккаунт, с которого идут запросы, может быть заблокирован.

tqdm - просто удобная вещь для отслеживания прогресса выполнения итераций цикла

```python
friends_my_friends = {}
bad_ids = []
for friend in tqdm(my_friends['items']):
    friend_id = friend['id']
    try:
        friends_my_friends[friend_id] = vk.friends.get(user_id=friend_id,
                                                       fields=['nickname', 'domain', 'sex', 'bdate', 'city', 
                                                               'country', 'timezone', 'photo_50', 'photo_100', 
                                                               'photo_200_orig', 'has_mobile', 'contacts', 
                                                               'education', 'online', 'relation', 'last_seen', 
                                                               'status', 'can_write_private_message', 
                                                               'can_see_all_posts', 'can_post', 'universities' ],)
    except Exception as e:
        bad_ids.append(friend_id)
    sleep(0.2)
print('Битые id: ', bad_ids)
```

Сохраним на диск

```python
#save pkl
with open('friends_my_friends.pkl', 'wb') as fp:
    pickle.dump(friends_my_friends, fp, protocol=pickle.HIGHEST_PROTOCOL)
    

#load pkl
#with open('friends_my_friends.pkl', 'rb') as fp:
#    data = pickle.load(fp)
```

## Постоение социального графа

Теперь создадим граф. Добавим вершины node и ребра edge. Код взят из <a href='https://habr.com/company/ods/blog/263313/'>статьи на Хабре</a>

```python
g = networkx.Graph(directed=False)
for i in friends_my_friends:
    g.add_node(i)
    for j in friends_my_friends[i]['items']:
        id_ = j['id']
        if i != id_ and i in friends_my_friends and id_ in friends_my_friends:
            g.add_edge(i, id_)
```

Вершины содержат только информацию об id соответствующего пользователя, но никак не привязаны к координатной плоскости. Функция <a href='https://networkx.github.io/documentation/networkx-1.11/reference/generated/networkx.drawing.layout.fruchterman_reingold_layout.html'>fruchterman_reingold_layout</a> сгенерирует координаты вершин графа.

```python
pos = networkx.fruchterman_reingold_layout(g, random_state=1) 

for k, v in list(pos.items())[:10]:
    pprint({k: v})
```

Зафиксируем порядок ключей в словаре. При работе с Python 2 (в Python 3.6 эта особенность <a href='https://docs.python.org/3/whatsnew/3.6.html#whatsnew36-compactdict'>отсутствует</a>) порядок элементов словаря будет меняться при каждом использовании словаря. Т.е.,например, при одном запуске на первом месте будет стоять вершина с ключом 310639377, при следующем запуске на первом может оказаться вершина с ключом 267824380. 

```python
pos_dict_keys = pos.keys()
```

Получим метки **Имя Фамилия** для всех друзей. Получение других данных (город, страна, место работы/учебы и т.д.) не гарантировано из-за закрытых профилей некоторых пользователей.

```python
id_name_dict = {item['id']: ' '.join([item['first_name'], item['last_name']]) for item in my_friends['items']}
del(id_name_dict[349376672])

labels = [id_name_dict[k] for k in pos_dict_keys]
labels[:10]
```

Назначим веса в зависимости от места в списке друзей. Чем выше, тем больше отметка на социальном графе. Первому другу будет соответствовать значение 20, последнему - 5

```python
N = np.linspace(0, 1, len(my_friends['items']))
id_weight_dict = {item['id']: 20 - 15*N[i] for i, item in enumerate(my_friends['items'])}

marker_size = [id_weight_dict[k] for k in pos_dict_keys]
marker_size[:10]
```

Теперь создадим списки координат вершин и ребер.

```python
Xv=[pos[k][0] for k in pos_dict_keys]
Yv=[pos[k][1] for k in pos_dict_keys]
Xed=[]
Yed=[]
for edge in g.edges():
    Xed += [pos[edge[0]][0], pos[edge[1]][0], None]
    Yed += [pos[edge[0]][1], pos[edge[1]][1], None] 
    
```

Создадим объекты Scatter - это линии и точки, которые будут отображены на графике. 

```python
edges = Scatter(x=Xed,
                y=Yed,
                mode='lines',
                line=dict(color='rgb(210,210,210)', width=1),
                hoverinfo='none'
               )
nodes = Scatter(x=Xv,
                y=Yv,
                mode='markers',
                marker=dict(symbol='circle-dot',
                              size=marker_size, 
                              color='#6959CD',
                              line=dict(color='rgb(50,50,50)', width=0.5)
                              ),
                text=labels,
                hoverinfo='text'
               )
```

Наконец нарисуем социальный граф.

```python
data = [edges, nodes]
figure = Figure(data=data, 
                layout=Layout(title='<br>Социальный граф ВК',
                              titlefont=dict(size=16),
                              showlegend=False,
                              hovermode='closest',
                              margin=dict(b=20,l=5,r=5,t=40),
                              annotations=[ dict(text=("При подговотке использовал код из " 
                                                       "<a href='https://habr.com/company/ods/blog/263313/'>статьи на Хабре</a> и "
                                                       "<a href='https://plot.ly/python/igraph-networkx-comparison/'>документацию Plotly</a>"),
                                                 showarrow=False,
                                                 xref="paper", yref="paper",
                                                 x=0.005, y=-0.002 
                                                ) 
                                          ],
                              xaxis=dict(showgrid=False, zeroline=False, showticklabels=False),
                              yaxis=dict(showgrid=False, zeroline=False, showticklabels=False)
                             )
)
iplot(figure)
```

## Кластеризация

Теперь попробуем найти кластеры в полученных данных

```python
points = np.array([[x, y] for x, y in pos.values()])
```

```python
tableau20 = [(31, 119, 180), (174, 199, 232), (255, 127, 14), (255, 187, 120),
             (44, 160, 44), (152, 223, 138), (214, 39, 40), (255, 152, 150),
             (148, 103, 189), (197, 176, 213), (140, 86, 75), (196, 156, 148),
             (227, 119, 194), (247, 182, 210), (127, 127, 127), (199, 199, 199),
             (188, 189, 34), (219, 219, 141), (23, 190, 207), (158, 218, 229)]
```

```python
num_clusters = 3
assert  num_clusters < len(tableau20), 'Недостаточно цветов для отображения всех кластеров'

#color_list = np.random.choice(tableau20, num_clusters)
random.shuffle(tableau20)
kmeans = KMeans(n_clusters=num_clusters, random_state=1).fit(points)
colors = ['rgb({0[0]},{0[1]},{0[2]})'.format(tableau20[pred]) for pred in kmeans.predict(points)]
```

```python
edges = Scatter(x=Xed,
                y=Yed,
                mode='lines',
                line=dict(color='rgb(210,210,210)', width=1),
                hoverinfo='none'
               )
nodes = Scatter(x=Xv,
                y=Yv,
                mode='markers',
                marker=dict(symbol='circle-dot',
                            size=marker_size, 
                            color=colors,
                            line=dict(color='rgb(50,50,50)', width=0.5),
                            opacity=0.9
                              ),
                text=labels,
                hoverinfo='text'
               )

data = [edges, nodes]
figure = Figure(data=data, 
                layout=Layout(title='<br>Социальный граф ВК',
                              titlefont=dict(size=16),
                              showlegend=False,
                              hovermode='closest',
                              margin=dict(b=20,l=5,r=5,t=40),
                              annotations=[ dict(text=("При подговотке использовал код из " 
                                                       "<a href='https://habr.com/company/ods/blog/263313/'>статьи на Хабре</a> и "
                                                       "<a href='https://plot.ly/python/igraph-networkx-comparison/'>документацию Plotly</a>"),
                                                 showarrow=False,
                                                 xref="paper", yref="paper",
                                                 x=0.005, y=-0.002 
                                                ) 
                                          ],
                              xaxis=dict(showgrid=False, zeroline=False, showticklabels=False),
                              yaxis=dict(showgrid=False, zeroline=False, showticklabels=False)
                             )
)
iplot(figure)
```
