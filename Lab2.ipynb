Импортируем необходимые библиотеки

```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

Определим функцию отрисовки графика

```python
def draw_plot(x, y, plot_size=(5,5)):
    plt.figure(figsize=plot_size)
    plt.plot(x,y,'-r', linewidth=3)
    plt.grid(True)
```

#### Уравнение прямой №1

$y = - \sqrt{9-(x + 6)^2}$

Определяем функцию прямой

```python
def f1(x):
    return -np.sqrt(9 - (x + 6) * (x + 6))
```


<p>Создаем массив точек для оси Х размером 100 в диапазоне от -4 до 0. Затем вычисляем значения функции для этого массива точек</p>


```python
x = np.linspace(-6, -9, 100)
y = f1(x)
```


<p>Строим график</p>


```python
draw_plot(x, y)
```

#### Уравнение окружности №2


$y = x + 3$

```python
def f2(x):
    return x + 3
```

```python
x = np.linspace(-6, -3, 100)
y = f2(x)
```

```python
draw_plot(x, y)
```

#### Уравнение окружности №3


$y = \sqrt{9-x^2}$

```python
def f3(x):
    return np.sqrt(9-x**2)
```

```python
x = np.linspace(-3, 0, 100)
y = f3(x)
```

```python
draw_plot(x, y)
```

#### Уравнение прямой №4


$y = -x + 3$

```python
def f4(x):
    return -x + 3
```

```python
x = np.linspace(0, 3, 100)
y = f4(x)
```

```python
draw_plot(x, y)
```

#### Уравнение прямой №5


$y = 0.5x - 1.5$

```python
def f5(x):
    return 0.5 * x - 1.5
```

```python
x = np.linspace(3, 9, 100)
y = f5(x)
```

```python
draw_plot(x, y, (10, 5))
```


<h4 id="Собираем-все-уравнения-в-одну-функцию">Собираем все уравнения в одну функцию<a class="anchor-link" href="#Собираем-все-уравнения-в-одну-функцию">¶</a></h4>


```python
def foo(x):
    if x >= -9 and x <= -6:
        return f1(x)
    elif x > -6 and x <= -3:
        return f2(x)
    elif x > -3 and x <= 0:
        return f3(x)
    elif x > 0 and x <= 3:
        return f4(x)
    elif x > 3 and x <= 9:
        return f5(x)
```

```python
x = np.linspace(-9, 9,1000)
y = [foo(i) for i in x]
```

```python
draw_plot(x, y, (18,6))
```
